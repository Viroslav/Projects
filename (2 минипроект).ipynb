Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

Проанализируйте даты запланированного прибытия. 
<br>– На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?
<br>– Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

```python
df.query('is_canceled == 1') \
    .query('arrival_date_year == 2015') \
    .groupby(['arrival_date_month', 'hotel']) \
    .agg({'arrival_date_month': 'count'}) \
    .arrival_date_month \
    .sort_values(ascending = False)
```

```python
df.query('is_canceled == 1') \
    .query('arrival_date_year == 2016') \
    .groupby(['arrival_date_month', 'hotel']) \
    .agg({'arrival_date_month': 'count'}) \
    .arrival_date_month \
    .sort_values(ascending = False)
```

Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
df.agg({'adults':'mean', 'children':'mean', 'babies':'mean'})
```

Создайте колонку total_kids, объединив children и babies. Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?

```python
df['total_kids'] = df.children + df.babies
```

```python
df.groupby('hotel').agg({'total_kids':'mean'}).round(2)
```

Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
df.total_kids > 0
```

```python
df['has_kids'] = df.total_kids > 0
```

```python
round((df.query('is_canceled == 1 and has_kids == False').shape[0] / df.query('has_kids == False').shape[0] * 100), 2) 
```

```python
kids_churn = df.query('is_canceled == 1 and has_kids == True').shape[0] / df.query('has_kids == True').shape[0]
kids_churn = round((kids_churn * 100),2)
kids_churn
```
